# 作業:畫函數圖形

- 笛卡爾心型線：$ F = X^2 + Y^2 + aX - a\sqrt{X^2 + Y^2} $
- 搭配塗色

```python
"""
繪製出笛卡爾心型線圖形，並搭配塗色。
"""

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


a = 0.5 # 設定參數 a，可影響心型的形狀
x = np.linspace(-2, 2, 400) # 建立 x 範圍
y = np.linspace(-2, 2, 400) # 建立 y 範圍
X, Y = np.meshgrid(x, y) # 建立 X，Y 網格點，形成二維平面


# 定義函數 F，這個函數決定了等高線的形狀
F = X**2 + Y**2 + a*X - a*np.sqrt(X**2 + Y**2)


plt.figure(figsize=(6,6)) # 使x軸和y軸比例一致


# 繪製等高線填色圖 (contourf)
# levels=50 代表將數值區間分成 50 個等級來填色，cmap="coolwarm" 設定顏色主題
plt.contourf(X, Y, F, levels=50, cmap="coolwarm")


# 繪製特定等高線 (contour)
# levels=[0, 1, 2] 代表我們只畫出 F=0、F=1、F=2 的等高線
plt.contour(X, Y, F, levels=[0, 1, 2], colors='black')


plt.grid(True) # 顯示格線
plt.show() # 顯示圖形
```
